Resumen de facturacion por tienda

```python
# Se crea diccionario y DataFrame con las 4 tiendas
datos = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Facturación': [facturacion_tienda, facturacion_tienda2, facturacion_tienda3, facturacion_tienda4]
}
df_resumen = pd.DataFrame(datos)

# Se calcula el total general
total_general = df_resumen['Facturación'].sum()

# Se agrega fila de total al DataFrame
df_resumen.loc[len(df_resumen)] = ['Total General', total_general]

# Formatear facturación con separador de miles
df_resumen['Facturación'] = df_resumen['Facturación'].map('{:,.0f}'.format)

# Mostrar tabla resumen
print("Tabla resumen de facturación por tienda:")
display(df_resumen)

# --- Gráfico de barras (sin el total general) ---
plt.figure(figsize=(9, 5))
plt.bar(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
        [facturacion_tienda, facturacion_tienda2, facturacion_tienda3, facturacion_tienda4],
        color=['skyblue', 'salmon', 'lightgreen', 'orange'])

plt.title('Facturación Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Facturación ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Agrupar y contar por Categoría para cada tienda, agregando total al final

# Tienda 1
categorias_tienda = tienda['Categoría del Producto'].value_counts().reset_index()
categorias_tienda.columns = ['Categoría', 'Cantidad']

# Agregar fila de total
total1 = pd.DataFrame([{'Categoría': 'TOTAL', 'Cantidad': categorias_tienda['Cantidad'].sum()}])
categorias_tienda1 = pd.concat([categorias_tienda, total1], ignore_index=True)

# Tienda 2
categorias_tienda2 = tienda2['Categoría del Producto'].value_counts().reset_index()
categorias_tienda2.columns = ['Categoría', 'Cantidad']

total2 = pd.DataFrame([{'Categoría': 'TOTAL', 'Cantidad': categorias_tienda2['Cantidad'].sum()}])
categorias_tienda2 = pd.concat([categorias_tienda2, total2], ignore_index=True)

# Tienda 3
categorias_tienda3 = tienda3['Categoría del Producto'].value_counts().reset_index()
categorias_tienda3.columns = ['Categoría', 'Cantidad']

total3 = pd.DataFrame([{'Categoría': 'TOTAL', 'Cantidad': categorias_tienda3['Cantidad'].sum()}])
categorias_tienda3 = pd.concat([categorias_tienda3, total3], ignore_index=True)

# Tienda 4
categorias_tienda4 = tienda4['Categoría del Producto'].value_counts().reset_index()
categorias_tienda4.columns = ['Categoría', 'Cantidad']

total4 = pd.DataFrame([{'Categoría': 'TOTAL', 'Cantidad': categorias_tienda4['Cantidad'].sum(),}])
categorias_tienda4 = pd.concat([categorias_tienda4, total4], ignore_index=True)

# Mostrar resultados
print("📌 Cantidad de productos vendidos por categoría - Tienda 1")
display(categorias_tienda1)

print("📌 Cantidad de productos vendidos por categoría - Tienda 2")
display(categorias_tienda2)

print("📌 Cantidad de productos vendidos por categoría - Tienda 3")
display(categorias_tienda3)

print("📌 Cantidad de productos vendidos por categoría - Tienda 4")
display(categorias_tienda4)
```

# 3. Calificación promedio de la tienda


```python
# Calcular calificación promedio por tienda
promedio_calificacion1 = tienda['Calificación'].mean()
promedio_calificacion2 = tienda2['Calificación'].mean()
promedio_calificacion3 = tienda3['Calificación'].mean()
promedio_calificacion4 = tienda4['Calificación'].mean()

# Crear un DataFrame con los resultados
calificaciones_promedio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [
        round(promedio_calificacion1, 2),
        round(promedio_calificacion2, 2),
        round(promedio_calificacion3, 2),
        round(promedio_calificacion4, 2)
    ]
})

# Mostrar el resultado
print("⭐ Calificación promedio por tienda:")
display(calificaciones_promedio)
```

```python
# Crear un gauge para cada tienda
fig = go.Figure()

for i, row in calificaciones_promedio.iterrows():
    fig.add_trace(go.Indicator(
        mode="gauge+number",
        value=row['Calificación Promedio'],
        title={'text': row['Tienda']},
        gauge={'axis': {'range': [0, 5]}},
        domain={'row': i // 2, 'column': i % 2}
    ))

fig.update_layout(
    grid={'rows': 2, 'columns': 2, 'pattern': "independent"},
    title="⭐ Calificación Promedio por Tienda"
)
fig.show()
```

# 4. Productos más y menos vendidos

```python
def analizar_top_bottom_tienda(df, nombre_tienda):
    # Contar productos vendidos
    ventas = df['Producto'].value_counts().reset_index()
    ventas.columns = ['Producto', 'Cantidad Vendida']

    top5 = ventas.head(5)
    bottom5 = ventas.tail(5)

    print(f"\n📍 {nombre_tienda}")
    print("🔝 Top 5 más vendidos:")
    print(top5.to_string(index=False))

    print("\n🔻 Bottom 5 menos vendidos:")
    print(bottom5.to_string(index=False))

    # Visualización conjunta en subplots
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    sns.barplot(data=top5, x='Cantidad Vendida', y='Producto', ax=axes[0], palette='crest')
    axes[0].set_title('Top 5 más vendidos')

    sns.barplot(data=bottom5, x='Cantidad Vendida', y='Producto', ax=axes[1], palette='rocket')
    axes[1].set_title('Bottom 5 menos vendidos')

    for ax in axes:
        ax.set_xlabel('Cantidad Vendida')
        ax.set_ylabel('Producto')

    fig.suptitle(f'Análisis de ventas en {nombre_tienda}', fontsize=14)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

```

```python
# Supongamos que tenés 4 DataFrames: tienda1, tienda2, tienda3, tienda4

analizar_top_bottom_tienda(tienda, "Tienda 1")
analizar_top_bottom_tienda(tienda2, "Tienda 2")
analizar_top_bottom_tienda(tienda3, "Tienda 3")
analizar_top_bottom_tienda(tienda4, "Tienda 4")

```

# 5. Envío promedio por tienda

```python
# Lista con tus DataFrames
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Lista para guardar resultados
costos_promedio_envio = []

# Calcular el costo promedio de envío por tienda
for df, nombre in zip(tiendas, nombres_tiendas):
    if 'Costo de envío' in df.columns:
        promedio_envio = df['Costo de envío'].mean()
        costos_promedio_envio.append({'Tienda': nombre, 'Costo Promedio de Envío': promedio_envio})
    else:
        print(f"⚠️ La columna 'Costo de envío' no existe en {nombre}")

# Convertir a DataFrame
df_envios_promedio = pd.DataFrame(costos_promedio_envio)
print(df_envios_promedio)
```

```python
fig = px.bar(df_envios_promedio,
             x='Tienda',
             y='Costo Promedio de Envío',
             text='Costo Promedio de Envío',
             title='📦 Costo Promedio de Envío por Tienda',
             color='Tienda',
             color_discrete_sequence=px.colors.qualitative.Vivid)

fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
fig.show()
```
